```python
def enropia(dcl):
    t = []
    if type(dcl) != type(t):
        dcl = dcl.to_list() # dcl = data cats list
        print('Now dcl is list')
    p1 = dcl.count(1)/len(dcl)
    p0 = 1 - p1 #dcl.count(0)/len(dcl)
    print('p1 = ', p1)
    print('p0 = ', p0)
    # E1 - for 0, E2 - for 1
    if p0 == 0:
        E0 = 0
    else:
        E0 =(p0*np.log2(p0))
    if p1 == 0:
        E1 = 0
    else:
        E1 = (p1*np.log2(p1))
    print('E0 = ', E0)
    print('E1 = ', E1)
    E = - E0 - E1
    print('E = ', E)
    return E

def slip_f(dc, y):
    dcl = dc.to_list()
    y_x = y.to_list()
    X_1 = [] # i > 1/2
    X_2 = [] # i < 1/2
    for i in range(len(dcl)):
        
        if dcl[i] > 0.5:
            X_1.append(y_x[i])
        elif dcl[i] < 1/2:
            X_2.append(y_x[i])
            
    return X_1, X_2        

def entr_p_m(X_1, X_2):
    print('X_1 = ', X_1)
    Ep = enropia(X_1)
    print('Ep = ', Ep)
    print('X_2 = ', X_2)
    Em = enropia(X_2)
    print('Em = ', Em)
    
    return Ep,Em
```

```python
Entropy_1 = entr_p_m(slip_f(dc1,dc4)[0],slip_f(dc1,dc4)[1])
Entropy_2 = entr_p_m(slip_f(dc2,dc4)[0],slip_f(dc2,dc4)[1])
Entropy_3 = entr_p_m(slip_f(dc3,dc4)[0],slip_f(dc3,dc4)[1])
```

```python
E_Y = enropia(dc4) # Энтропия исходная

def entr_y_x(dc, y): # Энтропия E(Y|X)
    #dcl = dc.to_list()
    y_x = y.to_list()
    p = slip_f(dc,y)[0]
    m = slip_f(dc,y)[1]
    n_1 = len(p)
    n_2 = len(m)
    N = len(y_x)
    E_p_m = entr_p_m(p,m)
    print('E_p_m = ', E_p_m)
    
    E = (n_1 / N * E_p_m[0]) - (n_2 / N * E_p_m[1])
    
    return E

E_Y_X1 = entr_y_x(dc1,dc4)
E_Y_X2 = entr_y_x(dc2,dc4)
E_Y_X3 = entr_y_x(dc3,dc4)

print('\n\nIG1 = ', E_Y - E_Y_X1)
print('\n\nIG2 = ', E_Y - E_Y_X2)
print('\n\nIG3 = ', E_Y - E_Y_X3)
```
